```python
### 6.SORU ###


from abc import ABC, abstractmethod

class Ozgecmis(ABC):
    @abstractmethod
    def egitim_bilgisi_ekle(self, okul_adi):
        pass

    @abstractmethod
    def is_deneyimi_ekle(self, sirket_adi):
        pass

class Calisan:
    def __init__(self, ad, soyad):
        self.ad = ad
        self.soyad = soyad
        self.egitim_bilgileri = []
        self.is_deneyimleri = []

    def calis(self):
        print(f"{self.ad} genel görevlerini yapıyor.")

    def bilgileri_goster(self):
        print(f"Ad: {self.ad}, Soyad: {self.soyad}")
        print(f"  Eğitim Bilgileri: {self.egitim_bilgileri}")
        print(f"  İş Deneyimleri: {self.is_deneyimleri}")

class MaviYaka(Calisan, Ozgecmis):
    def __init__(self, ad, soyad):
        super().__init__(ad, soyad)
        self.vardiya = 3

        print(f"{self.ad}, fabrikada tezgahta çalışıyor. (Vardiya: {self.vardiya})")

    def egitim_bilgisi_ekle(self, okul_adi):
        self.egitim_bilgileri.append(f"Okul: {okul_adi} (Mesleki Yeterlilik)")

    def is_deneyimi_ekle(self, sirket_adi):
        self.is_deneyimleri.append(f"Şirket: {sirket_adi} (Operatör)")

class BeyazYaka(Calisan, Ozgecmis):
    def __init__(self, ad, soyad):
        super().__init__(ad, soyad)
        self.vardiya = 2
    def calis(self):
        print(f"{self.ad}, ofiste masabaşı çalışıyor. (Vardiya: {self.vardiya})")
    def egitim_bilgisi_ekle(self, okul_adi):
        self.egitim_bilgileri.append(f"Okul: {okul_adi} (Lisans/Y.Lisans)")

    def is_deneyimi_ekle(self, sirket_adi):
        self.is_deneyimleri.append(f"Şirket: {sirket_adi} (Uzman/Yönetici)")

if __name__ == "__main__":
    print("--- Mavi Yaka Çalışan Oluşturuluyor ---")
    mavi_yaka_1 = MaviYaka("ELA", "AYDIN")
    mavi_yaka_1.egitim_bilgisi_ekle("Endüstri Meslek Lisesi")
    mavi_yaka_1.is_deneyimi_ekle("Demir Döküm A.Ş.")
    mavi_yaka_1.bilgileri_goster()

    print("\n" + "-"*30 + "\n")

    print("--- Beyaz Yaka Çalışan Oluşturuluyor ---")
    beyaz_yaka_1 = BeyazYaka("Ayşe", "Kaya")
    beyaz_yaka_1.egitim_bilgisi_ekle("İstanbul Medeniyet Üniversitesi")
    beyaz_yaka_1.is_deneyimi_ekle("Teknoloji Holding")
    beyaz_yaka_1.is_deneyimi_ekle("Yazılım Vadisi")

    beyaz_yaka_1.calis()
    beyaz_yaka_1.bilgileri_goster()
```
